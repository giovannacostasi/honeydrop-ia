Descrição: Este notebook apresenta a análise exploratória de dados de consumo alimentar de animais em situação de rua.

```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carregando o dataset
df = pd.read_csv('honeydrop_dataset.csv')

# Visualizando os dados
df.head()
```

```python
# Informações gerais
df.info()

# Estatísticas
df.describe()
```

```python
# Gráfico de consumo ao longo do tempo
plt.plot(df['hora'], df['peso_final_g'])
plt.xlabel('Hora')
plt.ylabel('Peso da Ração')
plt.title('Consumo de Ração ao Longo do Tempo')
plt.show()
```

```python
# Consumo por horário
plt.bar(df['hora'], df['consumo_g'])
plt.title('Consumo por Período')
plt.show()
```

A análise exploratória dos dados permitiu identificar padrões no consumo alimentar ao longo do tempo, com variações de acordo com o período do dia. Observa-se que há momentos de maior consumo, o que pode auxiliar na previsão de reabastecimento.
